# Experiment 66: Cluster Features

Notebook-обёртка для запуска `src/experiments_v2/66_cluster_features/run.py` из одного файла.

Что делает:
- находит корень проекта
- загружает `run.py` как модуль
- запускает `main()`
- показывает пути к артефактам и краткий summary

```python
from pathlib import Path
import os
import sys
import json
import importlib.util

NOTEBOOK_DIR = Path.cwd()
REPO_ROOT = NOTEBOOK_DIR
if not (REPO_ROOT / "src").exists():
    for parent in [NOTEBOOK_DIR, *NOTEBOOK_DIR.parents]:
        if (parent / "src").exists() and (parent / "data").exists():
            REPO_ROOT = parent
            break

os.chdir(REPO_ROOT)
if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))

RUN_PATH = REPO_ROOT / "src" / "experiments_v2" / "66_cluster_features" / "run.py"
METRICS_PATH = REPO_ROOT / "src" / "experiments_v2" / "66_cluster_features" / "metrics.json"
PRED_PATH = REPO_ROOT / "src" / "experiments_v2" / "66_cluster_features" / "predictions.csv"

print(f"Repo root: {REPO_ROOT}")
print(f"Run path:   {RUN_PATH}")
print(f"Metrics:    {METRICS_PATH}")
print(f"Preds:      {PRED_PATH}")
```

```python
spec = importlib.util.spec_from_file_location("exp66_run", RUN_PATH)
module = importlib.util.module_from_spec(spec)
assert spec.loader is not None
spec.loader.exec_module(module)

module.main()
```

```python
with open(METRICS_PATH, "r", encoding="utf-8") as f:
    metrics = json.load(f)

best = metrics["best_variant"]
best_result = metrics["results"][best]

print("Best variant:", best)
print("MAE:", best_result["mae"])
print("WMAPE:", best_result["wmape"])
print("R2:", best_result["r2"])
print("Test rows:", metrics["test_rows"])
print("Predictions file:", PRED_PATH)
```
